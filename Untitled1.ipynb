### **Data Description**

There are **196 observations** and **9 variables** in this data set, the varibales are: 

- **experience** *(character)* — describes the experience level of players. Including *Pro, Veteran, Amateur, Regular,* and *Beginner.*  
- **subscribe** *(logical)* — indicates whether the player subscribed to game-related information (*TRUE* or *FALSE*).  
- **hashedEmail** *(character)* — this records the hashing encoded email addresses of players.  
- **played_hours** *(double)* — the total time a player spent in the game (in hours), ranging from *0 to 223.1.*  
- **name** *(character)* — player's name.  
- **gender** *(character)* — self-reported gender (e.g., Male, Female, Non-binary, etc.).  
- **age** *(double)* — player’s age, ranging from *8 to 99.*  
- **individualId** *(logical)* — mostly missing (NA) values.  
- **organizationName** *(logical)* — mostly missing (NA) values.

## Exploratory Data Analysis
Below are some summaries and plots to better understand the data.

```r
ggplot(players, aes(x = subscribe)) +
  geom_bar() +
  facet_grid(cols = vars(experience))
```

**Observation:**  
Most *Amateur*, *Beginner*, and *Veteran* players have higher subscription counts compared to *Pro* and *Regular* players.  
This suggests that experience level may influence subscription behavior, but the relationship is not strictly increasing with experience.

```r
ggplot(players, aes(x = played_hours)) +
  geom_histogram(bins = 30) +
  facet_grid(cols = vars(subscribe)) +
  labs(title = "Distribution of Played Hours by Subscribe Status",
       x = "Played Hours", y = "Count")
```

**Observation:**  
Most players play less than 20 hours, and only a few play more than 50 hours.  
The data are right-skewed because a small number of players play much longer than others.


```r
players |>
  filter(played_hours <= 50, Age <= 50) |>
  ggplot(aes(x = played_hours, y = Age, color = subscribe)) +
  geom_point() +
  labs(title = "Age vs Played Hours",
       x = "Played Hours", y = "Age")
```

**Observation:**  
There is no clear relationship between age and hours played. And both subscribed and non-subscribed players appear across all age groups.


```r
ggplot(players, aes(x = experience, y = played_hours)) +
  geom_point(alpha = 0.3) +
  labs(title = "Played Hours by Experience Level",
       x = "Experience Level", y = "Played Hours")
```

**Observation:**  
*Pro* and *Veteran* players usually have more played hours than *Beginner* and *Regular* players.  
This means experience level is likely related to total playtime.


```r
**Summary**
- The data are right-skewed with a few heavy players.  
- Higher experience seems linked to higher playtime and subscription.  
- Age shows little relationship with activity.  
These insights support using **experience**, **gender**, and **age** to predict **played_hours** in later modeling.
```
